Q1a:
The five methods and/or attributes that can be used to explore a pandas DataFrame or Series are as follows:

1. **sort_values():** The sort_values() method is used to sort the values in a data structure along a specified axis, either by column or index. For example, df['column_name'].sort_values() sorts the DataFrame based on the values in the specified column.

2. **value_counts():** The value_counts() method is used to count the occurrences of unique values in a data structure, and is particularly useful for exploring categorical data. For example, df['column_name'].value_counts() provides a count of unique values in the specified column of the DataFrame.

3. **head():** The head() method is used to display the first few elements of a data structure. For example, df.head(10) displays the first 10 rows of the DataFrame.

4. **unique():** The unique() method returns an array of unique values in a DataFrame or another data structure. For example, df['column_name'].unique() provides an array of unique values in the specified column of the DataFrame.

5. **dtype:** The dtype attribute returns the data type of the elements in a data structure. For example, df['column_name'].dtype returns the data type of the elements in the specified column of the DataFrame.

```python
def create_dataframe(num_rows, num_cols, col_names=None, row_index=None, rand_seed=12345):
    '''
    Create a DataFrame with specified dimensions, column names, row indices, and optional random data.
        Parameters:
        - num_rows (int): Number of rows in the DataFrame.
        - num_cols (int): Number of columns in the DataFrame.
        - col_names (list or None): List of column names. If None, default column names ('A', 'B', ..., 'Z') will be used.
        - row_index (list or None): List of row indices. If None, default row indices (0, 1, ..., num_rows-1) will be used.
        - rand_seed (int or None): Random seed for generating random data. If None, the DataFrame will be filled with NaN values.
        Returns:
        - DataFrame: A pandas DataFrame with the specified dimensions, column names, row indices, and optional random data.
    '''
    # Check if column names are provided; if not, use default column names
    # Code generated by Copilot
    if col_names is None:
        col_names = list('ABCDEFGHIJKLMNOPQRSTUVWXYZ')
    # Check if row index values are provided; if not, use default row indices
    # Code generated by Copilot
    if row_index is None:
        row_index = np.arange(num_rows)
    # Check if a random seed is provided; if yes, set the random seed
    # Code generated by Copilot
    if rand_seed is not None:
        np.random.seed(rand_seed)
        # Generate random values between -1 and 1
        data = np.random.uniform(low=-1, high=1, size=(num_rows, num_cols))
    else:
        # If no random seed, fill the DataFrame with NaN values
        data = np.full((num_rows, num_cols), np.nan)
    # Create and return a DataFrame with the specified data, columns, and index
    return pd.DataFrame(data, columns=col_names, index=row_index)
```

```python
# Using the create_dataframe function to create the DataFrame
df = create_dataframe(num_rows=3, num_cols=3, col_names=['a', 'b', 'c'], row_index=['x', 'y', 'z'])

# Method 1: sort_values()
sorted_df = df.sort_values(by='a')
print("\nMethod 1: sort_values()")
print(sorted_df)
print(f"Explanation 1: The sort_values() method sorts the DataFrame based on the values in the 'a' column, allowing us to see the DataFrame in ascending order of column 'a'.")

# Method 2: value_counts()
value_counts_col_a = df['a'].value_counts()
print("\nMethod 2: value_counts()")
print(value_counts_col_a)
print(f"Explanation 2: The value_counts() method provides a count of unique values in the 'a' column, helping us understand the frequency distribution of values in this column.")

# Method 3: head()
print("\nMethod 3: head()")
print(df.head())
print(f"Explanation 3: The head() method displays the first few rows of the DataFrame, giving us a quick overview of the structure and content of the DataFrame.")

# Method 4: unique()
unique_values_col_b = df['b'].unique()
print("\nMethod 4: unique()")
print(unique_values_col_b)
print(f"Explanation 4: The unique() method returns an array of unique values in the 'b' column, providing insight into the distinct values present in this column.")

# Attribute 5: dtype
dtype_col_c = df['c'].dtype
print("\nAttribute 5: dtype")
print(dtype_col_c)
print(f"Explanation 5: The dtype attribute returns the data type of the elements in the 'c' column, allowing us to understand the type of data stored in this column.")
```

### Q5: 15 pt:
[Blood pressure](https://www.health.harvard.edu/heart-health/reading-the-new-blood-pressure-guidelines) is reported as two numbers: diastolic (the lower number) and systolic (the higher number).  So, for example, if a person has a diastolic measurement of 80 and a systolic measurement of 120, the person's blood pressure would be reported as 120/80.

Blood pressure measurements fall into several categories:

Normal blood pressure. Your blood pressure is normal if it's below 120/80 mm Hg.

Elevated blood pressure. Elevated blood pressure is a systolic pressure ranging from 120 to 129 mm Hg and a diastolic pressure below (not above) 80 mm Hg. Elevated blood pressure may also be called prehypertension.

Stage 1 hypertension. Stage 1 hypertension is a systolic pressure ranging from 130 to 139 mm Hg or a diastolic pressure ranging from 80 to 89 mm Hg.

Stage 2 hypertension. More-severe hypertension, stage 2 hypertension is a systolic pressure of 140 mm Hg or higher or a diastolic pressure of 90 mm Hg or higher.

Hypertensive crisis. A blood pressure measurement higher than 180/120 mm Hg is an emergency situation that requires urgent medical care.

a) Report the number of people in each of those categories.

b) The guidelines were recently changed: prior to 2017, the guidelines set the threshold at 140/90 mm Hg for people younger than age 65 and 150/80 mm Hg for those ages 65 and older to be diagnosed with hypertension. Create and implement analyses that allow you to demonstrate how the change in guidelines affects the categorization of people into the different categories.
Make reasonable assumptions (and state them) about whether to use diastolic, systolic, both, or either of those measures when implementing your analysis.  Note that the new guidelines offer more categories than the older guidelines.

```python
# Function to categorize blood pressure under both new guidelines
def categorize_blood_pressure(systolic, diastolic):
    '''
    Categorize blood pressure based on systolic and diastolic measurements.
        Parameters:
        - systolic (float): The systolic blood pressure measurement.
        - diastolic (float): The diastolic blood pressure measurement.
        Returns:
        - str: A string indicating the blood pressure category.
    '''
    # Normal blood pressure category
    if systolic < 120 and diastolic < 80:
        return 'Normal blood pressure'
    # Elevated blood pressure (Prehypertension) category
    elif systolic >= 120 and systolic <= 129 and diastolic < 80:
        return 'Elevated blood pressure (Prehypertension)'
    # High blood pressure (Stage 1 Hypertension) category
    elif systolic >= 130 and systolic <= 139 or diastolic >= 80 and diastolic <= 89:
        return 'High blood pressure (Stage 1 Hypertension)'
    # High blood pressure (Stage 2 Hypertension) category
    elif systolic >= 140 or diastolic >= 90:
        return 'High blood pressure (Stage 2 Hypertension)'
    # Hypertensive Crisis category
    elif systolic >= 180 or diastolic >= 120:
        return 'Hypertensive Crisis (Urgent medical care needed)'

# Function to categorize blood pressure under both old guidelines
def categorize_blood_pressure_old(systolic, diastolic, age):
    '''
    Categorize blood pressure based on systolic and diastolic measurements and age.
        Parameters:
        - systolic (float): The systolic blood pressure measurement.
        - diastolic (float): The diastolic blood pressure measurement.
        - age (int): The age of the patient.
        Returns:
        - str: A string indicating the blood pressure category.
    '''
    # Check if age is below 65
    if age < 65:
        # Hypertension category for systolic >= 140 or diastolic >= 90
        if systolic >= 140 or diastolic >= 90:
            return 'Hypertension'
        # Normal blood pressure category for other cases
        else:
            return 'Normal blood pressure'
    else:
        # Hypertension category for systolic >= 150 or diastolic >= 80
        if systolic >= 150 or diastolic >= 80:
            return 'Hypertension'
        # Normal blood pressure category for other cases
        else:
            return 'Normal blood pressure'

# Read the CSV file into a DataFrame
df = pd.read_csv('/Users/anupamabhatta/Desktop/pwd/SI 618/SI_618_WN_24_Files/data/bodyPerformance.csv')

# Create a new column 'blood_pressure_category' using the categorize_blood_pressure function
# Generated by Copilot
df['blood_pressure_category'] = df.apply(lambda row: categorize_blood_pressure(row['systolic'], row['diastolic']), axis=1)

# Create a new column 'blood_pressure_category_old' using the categorize_blood_pressure_old function
df['blood_pressure_category_old'] = df.apply(lambda row: categorize_blood_pressure_old(row['systolic'], row['diastolic'], row['age']), axis=1)

# Display the counts for each blood pressure category under new as well as old guidelines
print("New Guidelines:")
print(df['blood_pressure_category'].value_counts())

print("\nOld Guidelines:")
print(df['blood_pressure_category_old'].value_counts())
```

**Analysis & Recommendation:**
The blood pressure analysis in the dataset compares two sets of guidelines: the new guidelines, which offer detailed classifications based on both systolic and diastolic measurements (Normal, Elevated, Stage 1 Hypertension, Stage 2 Hypertension, and Hypertensive Crisis), and the older guidelines, with broader categories ('Normal blood pressure' and 'Hypertension') depending on age and either systolic or diastolic values. The new guidelines provide a more accurate and nuanced understanding, aligning with modern medical standards. The older guidelines, on the other hand, are less precise due to simplicity and age-specific thresholds. For comprehensive health assessments, it is definitely recommended to follow the new guidelines, considering both systolic and diastolic measurements for a detailed and accurate classification in line with contemporary medical practices.

Implement a function to calculate each person's Body Mass Index (BMI). Use that function to create a new Series called 'BMI' in your DataFrame.

> With the metric system, the formula for BMI is weight in kilograms divided by height in meters squared. Since height is commonly measured in centimeters, an alternate calculation formula, dividing the weight in kilograms by the height in centimeters squared, and then multiplying the result by 10,000, can be used. (https://www.cdc.gov/nccdphp/dnpao/growthcharts/training/bmiage/page5_1.html)

Using the pandas-matplotlib integration (i.e. do not use other visualization libraries), create two histograms of BMI: one for gender F and one for gender M.  Describe similarities and differences of the two histograms.

```python
# Read the CSV file into a DataFrame
df = pd.read_csv('/Users/anupamabhatta/Desktop/pwd/SI 618/SI_618_WN_24_Files/data/bodyPerformance.csv')

# Function to calculate BMI
def calculate_bmi(weight, height):
    '''
    Calculate BMI using weight and height measurements.
        Parameters:
        - weight (float): The weight of the patient in kilograms.
        - height (float): The height of the patient in centimeters.
        Returns:
        - float: The BMI of the patient.
    '''
    return round((weight / (height ** 2)) * 10000, 2)

# Create a new column 'BMI' using the calculate_bmi function
# Generated by Copilot
df['BMI'] = df.apply(lambda row: calculate_bmi(row['weight_kg'], row['height_cm']), axis=1)

# Create histograms for BMI by gender
plt.figure(figsize=(8, 6))
df[df['gender'] == 'F']['BMI'].plot(kind='hist', color='pink', alpha=0.8, label='Female', legend=True)
df[df['gender'] == 'M']['BMI'].plot(kind='hist', color='blue', alpha=0.5, label='Male', legend=True)

plt.title('BMI Distribution by Gender')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.show()
```

### Q9: 18 pt:
Take a look at gripForce (a measure of strength), "sit and bend forward" (a measure of flexibility), and "sit-ups count" (a measures of strength and endurance of the abdominal and hip-flexor muscles).  For each of those measures, calculate their z-scores (i.e. subtract the mean of the variable from each observation, then divide by the standard deviation of that score).  Create an overall fitness score by adding the z-scores for each of those variables together, then use `pd.qcut()` (not `pd.cut()`) to assign them to four categories ranging from lowest score to highest score. As with previous questions, you should use one or more functions to implement your solution.  You should also include a written summary of your findings.  Be sure to include units in your written summary.

a) Report the number of individuals in each of the resulting categories.

b) Comment on the degree to which your categories correspond to the `class` column.

```python
# Read the CSV file into a DataFrame
df = pd.read_csv('/Users/anupamabhatta/Desktop/pwd/SI 618/SI_618_WN_24_Files/data/bodyPerformance.csv')

# Function to calculate z-scores and create overall fitness score
def calculate_overall_fitness_score(df):
    '''
    Calculate z-scores for grip force, sit and bend forward, and sit-ups counts, and create an overall fitness score.
        Parameters:
        - df (DataFrame): The DataFrame containing the data.
    '''
    # Calculate z-scores for each variable
    df['gripForce_z-score'] = (df['gripForce'] - df['gripForce'].mean()) / df['gripForce'].std()
    df['sit_and_bend_forward_z-score'] = (df['sit and bend forward_cm'] - df['sit and bend forward_cm'].mean()) / df['sit and bend forward_cm'].std()
    df['sit-ups_count_z-score'] = (df['sit-ups counts'] - df['sit-ups counts'].mean()) / df['sit-ups counts'].std()

    # Create overall fitness score
    df['overall_fitness_score'] = df['gripForce_z-score'] + df['sit_and_bend_forward_z-score'] + df['sit-ups_count_z-score']

    # Categorize using pd.qcut
    df['fitness_category'] = pd.qcut(df['overall_fitness_score'], q=[0, 0.25, 0.5, 0.75, 1], labels=['Low', 'Moderate', 'High', 'Very High'])

# Call the function to calculate fitness scores and categories
calculate_overall_fitness_score(df)

# Display the number of individuals in each category
category_total = df['fitness_category'].value_counts()
print(category_total)

# Display the correspondence between fitness categories and classes
class_correspondence = df.groupby('class')['fitness_category'].value_counts()
print(class_correspondence)
```

**Summary:**
In the code above, z-scores were calculated for three fitness measures: gripForce (a measure of strength), sit and bend forward_cm (a measure of flexibility), and sit-ups counts (a measure of strength and endurance of the abdominal and hip-flexor muscles). These z-scores were then used to create an overall fitness score for each individual. The overall fitness scores were categorized into four groups (Low, Moderate, High, Very High) using pd.qcut(). Examining the correspondence with the 'class' column, it is observed that different classes have varying proportions of individuals in each fitness category. For instance, individuals in class A have a higher representation in the 'Very High' and 'High' fitness categories, while class D has a larger proportion in the 'Low' fitness category. These findings suggest that the overall fitness score, derived from a combination of strength, flexibility, and endurance measures, shows variability across different classes.
